```python
driver = webdriver.Chrome()
#   2013_06_24_2019_06_30
#   https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term=%22artificial+intelligence%22&terms-0-field=title&terms-1-operator=OR&terms-1-term=AI&terms-1-field=title&terms-2-operator=OR&terms-2-term=%22deep+learning%22&terms-2-field=title&terms-3-operator=OR&terms-3-term=%22artificial+intelligence%22&terms-3-field=abstract&terms-4-operator=OR&terms-4-term=AI&terms-4-field=abstract&terms-5-operator=OR&terms-5-term=%22deep+learning%22&terms-5-field=abstract&classification-physics_archives=all&classification-include_cross_list=include&date-year=&date-filter_by=date_range&date-from_date=2013-06-24&date-to_date=2019-06-30&date-date_type=submitted_date&abstracts=show&size=200&order=-announced_date_first

#   2019_07_01_2020_10_31
#   https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term=%22artificial+intelligence%22&terms-0-field=title&terms-1-operator=OR&terms-1-term=AI&terms-1-field=title&terms-2-operator=OR&terms-2-term=%22deep+learning%22&terms-2-field=title&terms-3-operator=OR&terms-3-term=%22artificial+intelligence%22&terms-3-field=abstract&terms-4-operator=OR&terms-4-term=AI&terms-4-field=abstract&terms-5-operator=OR&terms-5-term=%22deep+learning%22&terms-5-field=abstract&classification-physics_archives=all&classification-include_cross_list=include&date-year=&date-filter_by=date_range&date-from_date=2019-07-01&date-to_date=2020-10-31&date-date_type=submitted_date&abstracts=show&size=200&order=-announced_date_first

#   2020_11_01_2021_10_31
#   https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term=%22artificial+intelligence%22&terms-0-field=title&terms-1-operator=OR&terms-1-term=AI&terms-1-field=title&terms-2-operator=OR&terms-2-term=%22deep+learning%22&terms-2-field=title&terms-3-operator=OR&terms-3-term=%22artificial+intelligence%22&terms-3-field=abstract&terms-4-operator=OR&terms-4-term=AI&terms-4-field=abstract&terms-5-operator=OR&terms-5-term=%22deep+learning%22&terms-5-field=abstract&classification-physics_archives=all&classification-include_cross_list=include&date-year=&date-filter_by=date_range&date-from_date=2020-11-01&date-to_date=2021-10-31&date-date_type=submitted_date&abstracts=show&size=200&order=-announced_date_first

#   2021_11_01_2022_09_30
#   https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term=%22artificial+intelligence%22&terms-0-field=title&terms-1-operator=OR&terms-1-term=AI&terms-1-field=title&terms-2-operator=OR&terms-2-term=%22deep+learning%22&terms-2-field=title&terms-3-operator=OR&terms-3-term=%22artificial+intelligence%22&terms-3-field=abstract&terms-4-operator=OR&terms-4-term=AI&terms-4-field=abstract&terms-5-operator=OR&terms-5-term=%22deep+learning%22&terms-5-field=abstract&classification-physics_archives=all&classification-include_cross_list=include&date-year=&date-filter_by=date_range&date-from_date=2021-11-01&date-to_date=2022-09-30&date-date_type=submitted_date&abstracts=show&size=200&order=-announced_date_first

#   2022_10_01_2023_06_30
#   https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term=%22artificial+intelligence%22&terms-0-field=title&terms-1-operator=OR&terms-1-term=AI&terms-1-field=title&terms-2-operator=OR&terms-2-term=%22deep+learning%22&terms-2-field=title&terms-3-operator=OR&terms-3-term=%22artificial+intelligence%22&terms-3-field=abstract&terms-4-operator=OR&terms-4-term=AI&terms-4-field=abstract&terms-5-operator=OR&terms-5-term=%22deep+learning%22&terms-5-field=abstract&classification-physics_archives=all&classification-include_cross_list=include&date-year=&date-filter_by=date_range&date-from_date=2022-10-01&date-to_date=2023-06-30&date-date_type=submitted_date&abstracts=show&size=200&order=-announced_date_first

#   2023_07_01_2024_01_31
#   https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term=%22artificial+intelligence%22&terms-0-field=title&terms-1-operator=OR&terms-1-term=AI&terms-1-field=title&terms-2-operator=OR&terms-2-term=%22deep+learning%22&terms-2-field=title&terms-3-operator=OR&terms-3-term=%22artificial+intelligence%22&terms-3-field=abstract&terms-4-operator=OR&terms-4-term=AI&terms-4-field=abstract&terms-5-operator=OR&terms-5-term=%22deep+learning%22&terms-5-field=abstract&classification-physics_archives=all&classification-include_cross_list=include&date-year=&date-filter_by=date_range&date-from_date=2023-07-01&date-to_date=2024-01-31&date-date_type=submitted_date&abstracts=show&size=200&order=-announced_date_first

#   2024_02_01_2024_06_24
#   https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term=%22artificial+intelligence%22&terms-0-field=title&terms-1-operator=OR&terms-1-term=AI&terms-1-field=title&terms-2-operator=OR&terms-2-term=%22deep+learning%22&terms-2-field=title&terms-3-operator=OR&terms-3-term=%22artificial+intelligence%22&terms-3-field=abstract&terms-4-operator=OR&terms-4-term=AI&terms-4-field=abstract&terms-5-operator=OR&terms-5-term=%22deep+learning%22&terms-5-field=abstract&classification-physics_archives=all&classification-include_cross_list=include&date-year=&date-filter_by=date_range&date-from_date=2024-01-31&date-to_date=2024-06-24&date-date_type=submitted_date&abstracts=show&size=200&order=-announced_date_first

url = "https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term=%22artificial+intelligence%22&terms-0-field=title&terms-1-operator=OR&terms-1-term=AI&terms-1-field=title&terms-2-operator=OR&terms-2-term=%22deep+learning%22&terms-2-field=title&terms-3-operator=OR&terms-3-term=%22artificial+intelligence%22&terms-3-field=abstract&terms-4-operator=OR&terms-4-term=AI&terms-4-field=abstract&terms-5-operator=OR&terms-5-term=%22deep+learning%22&terms-5-field=abstract&classification-physics_archives=all&classification-include_cross_list=include&date-year=&date-filter_by=date_range&date-from_date=2024-02-01&date-to_date=2024-06-24&date-date_type=submitted_date&abstracts=show&size=200&order=-announced_date_first"

driver.get(url)

next_page_num = 0

# 상단에 있는 페이지 번호
page_ul_tag = driver.find_element(By.CSS_SELECTOR, '.pagination-list')

# 현재 페이지에 표시되어있는 페이지 번호들
page_li_tag = page_ul_tag.find_elements(By.CSS_SELECTOR, 'li a.pagination-link')

df = pd.DataFrame(columns=['title','pdf_link', 'submit_date', 'Authors' ,'Abstract'])

while int(driver.find_element(By.CSS_SELECTOR,'.pagination-list .is-current').text) < 51 :
    curr_page_num = driver.find_element(By.CSS_SELECTOR,'.pagination-list .is-current').text

    # 페이지가 1페이지만 있는게 아닐 경우
    if len(page_li_tag) != 1 :
        curr_page_num

    new_df = pd.DataFrame(columns=['title','pdf_link', 'submit_date', 'Authors' ,'Abstract'])

    # 자바스크립트 활용해서 요약 전부 펼치기
    js_script = """
        document.querySelectorAll('.abstract-full').forEach(function(element) {
            element.style.display = 'inline';
        });
    """
    driver.execute_script(js_script)

    # 
    ol_tag = driver.find_element(By.CSS_SELECTOR,'ol.breathe-horizontal')

    li_tags = ol_tag.find_elements(By.TAG_NAME,'li')

    # 제목
    title_list = []
    for li_tag in li_tags :
        title_list.append(li_tag.find_element(By.CSS_SELECTOR, 'p.title.is-5.mathjax').text)

    # 작성자
    authors_list = []
    for li_tag in li_tags :
        authors_list.append(li_tag.find_element(By.CSS_SELECTOR, 'p.authors a').text)

    # 제출 날짜
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    submit_date_list = []
    for li_tag in li_tags :

        # Submitted 11 April, 2024; originally announced July 2024. ;로 분리
        split_texts = li_tag.find_element(By.CSS_SELECTOR, 'p.is-size-7').text.split(';')
        
        # 첫번째 문자들에서 'Submitted', ',' 제거 후 공백으로 분리
        date = re.sub(r'(Submitted|\,)', '',split_texts[0]).split()

        # 영어 월 가져와서 숫자로 변환 해주기
        en_month = date[1]
        month = ""
        if en_month in months :
            month = months.index(en_month) + 1  # list index가 0부터 시작이라 + 1 해주기
        
        # 10월이 안될경우 앞에 0 붙여주기
        if month < 10 :
            month = '0' + str(month)
        
        day = ""
        if(int(date[0]) < 10) :
            day = "0" + str(date[0])
        else :
            day = str(date[0])
        
        submit_date = date[2] + '-' + str(month) + '-' + day
        
        submit_date_list.append(submit_date)
    
    # 요약내용
    abstract_list = []
    for li_tag in li_tags :
        abstract_txt = li_tag.find_element(By.CSS_SELECTOR, 'p.abstract.mathjax span.abstract-full').text
        abstract_list.append(re.sub(r'(▽ More|△ Less)', '',abstract_txt))


    # pdf link
    pdf_link_list = []
    for idx,li_tag in enumerate(li_tags) :
        pdf_link = ""
        try :
            pdf_link = li_tag.find_element(By.CSS_SELECTOR, 'p.list-title.is-inline-block span a').get_attribute('href')
        except Exception as e:
            print(idx)
            
        pdf_link_list.append(pdf_link)

    'title','authors','submit_date','pdf_link','Abstract'
    new_df['title'] = title_list
    new_df['Authors'] = authors_list
    new_df['submit_date'] = submit_date_list
    new_df['pdf_link'] = pdf_link_list
    new_df['Abstract'] = abstract_list

    lenth_df = len(df)

    df = pd.concat([df, new_df], ignore_index=True)

    # 현재 담아놓은 데이터 갯수 비교
    while len(df) != (len(new_df) + lenth_df) :
        print(f'len(df) : {len(df)}')
        print(f'(len(new_df) + lenth_df) : {(len(new_df) + lenth_df)}')
        print('데이터 갯수 비교 중')
        time.sleep(1)
    

    curr_page_num = driver.find_element(By.CSS_SELECTOR,'.pagination-list .is-current').text    # 현재 페이지 번호
    next_page_num = int(curr_page_num) + 1  # 다음 페이지 번호
    
    try :
        # 다음페이지 이동 버튼
        next_btn = driver.find_element(By.CSS_SELECTOR,'nav a.pagination-next')

        # 마지막 페이지로 갈 경우 is-invisible 클래스를 추가해서 next버튼을 숨겨주기때문에 마지막 페이지일 경우에는 멈춘다
        if (next_btn.get_attribute('class') in 'is-invisible') or (next_btn.get_attribute('disabled')):
            break
        
        # 다음페이지 이동
        next_btn.click()
    except Exception as e :
        # 다음페이지 버튼이 없을경우 종료
        break        

    # 현재 페이지 번호가 다음페이지 번호가 아닐 경우 대기
    while driver.find_element(By.CSS_SELECTOR,'.pagination-list .is-current').text != str(next_page_num) :
        print('페이지 이동 중')
        time.sleep(1)
    
df.to_excel('arxiv_2024_02_01_2024_06_24.xlsx', index=False)
```
