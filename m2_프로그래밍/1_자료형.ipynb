```python
import keyword

strings = ["var1","2things","variable_name","time!"]

def identifyKey(strings):

  count = 0
  # 키워드사용안됨
  for i in range(len(strings)):
    print(i)
    print(strings[i][0])

    # 넣은 변수와 키워드 비교 후 없으면 리턴
    if i not in keyword.kwlist:
      print(i)
    # 있으면
      break
      print('Not Available')

  # 첫 인덱스에 숫자난 공백 사용 불가
  # 첫 인덱스가 0부터 9인 숫자인 경우 제거
  for j in range(10):
    if strings[i][0] == str(j):
      print(strings[i][0])
      i -= 1
      # 해당 인덱스 제거
      strings.pop(i)

      print('ok')

  # 언더바 외 특수문자와 공백 사용 불가
  for k in strings:
    print(strings[i][k])
    count += 1
    print(k[i])
    print(strings[i][k])
    count += 1
    print(k[i])
    print(strings[k])
    # 각 리스트의 인덱스에 특수문자와 공백 있을 시 제거
    #if k[i] == '!':
    if i[j] == '@' or i[j] == '#' or i[j] == '%' or i[j] == '^' or i[j] == '*' or i[j] == ' ':
      strings.pop(k)

  # 결과 값 리턴
  return strings

print(identifyKey(identifiers))
```

1. 메서드 vs 함수:
- sort(): 리스트 객체의 내장 메서드입니다. 즉, 리스트에서만 사용할 수 있습니다.
- sorted(): 내장 함수로, 어떤 반복 가능한(iterable) 객체에도 사용될 수 있습니다. 예를 들면 리스트, 튜플, 딕셔너리, 문자열 등에 사용할 수 있습니다.

2. 반환 값:
- sort(): 리스트를 원 위치에서(in-place) 정렬하고 None을 반환합니다. 따라서 원래의 리스트 자체가 변경됩니다.
- sorted(): 정렬된 새로운 리스트를 반환합니다. 원래의 객체는 변경되지 않습니다.

3. 유용성:
- sort(): 리스트에서만 작동하기 때문에 리스트만 정렬할 수 있습니다.
- sorted(): 다양한 객체를 정렬할 수 있으며 결과는 항상 리스트로 반환됩니다.

```python
# sort()
my_list = [3, 1, 2]
my_list.sort()
print(my_list)

# sorted()
my_tuple = (3, 1, 2)
new_list = sorted(my_tuple)
print(my_tuple)
print(new_list)
```

```python
# list.sort() -> sort()는 list자체에서 정렬시키는거기 때문에 list안에 값의 순서가 바뀌고 return되는 값이 없다

list = [5,6,9,7,8]
new_list = list.sort()
print(list)
print(new_list)
```

```python
# sorted -> 원래 객체의 값은 건드리지않고 정렬된값을 return 시켜주기때문에 원래 객체는 처음 그대로이다

list = [5,6,9,7,8]
new_list = sorted(list)
print(list)
print(new_list)
```

```python
# 위치 좌표의 리스트
positions = [(1,1),(2,2),(1,1),(3,3),(2,2)]

# 각 위치의 출현 횟수를 저장할 딕셔너리
position_count = {}

for pos in positions:
    if pos in position_count:
      position_count[pos] += 1
    else :
      position_count[pos] = 1

print(position_count)
```

예시 출력: [4, 5]

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

#print(set(list1)&set(list2))

# &:교집합, |:합집합

def intersection_list(l1,l2) :
  if not l1 or not l2:
    return l1,l2
  result = set(l1)&set(l2)

  return list(result)

print(intersection_list(list1,list2))
```

[ 예시 ]

문장을 입력하세요: I love apple. I love orange. Apple is tasty

{'i': 2, 'love': 2, 'apple.': 1, 'orange.': 1, 'apple': 1, 'is': 1, 'tasty': 1}

```python
import re
#str = "I love apple. I love orange. Apple is tasty"
str = input()

# 문자를 남기기 위해 특수문자 제거
str = re.sub("[!@#$%^&*()<>.]", "", str)

# 특수문자 제거 한 문자에 대소문자가 섞여있을 수가 있어 lower를 사용하여 모두 소문자로 변환시켜준다
# 대문자로 변환하는거는 upper
str = str.lower()

#print("특수문자 제거 : " , re.sub("[!@#$%^&*()<>.]","",str))

# 특수문자를 제거 하고 난 후에 문자를 단어 단위로 분리
split_str = str.split()
#print(str.split())

stri = {}

for i in split_str:
  if i in stri:
    stri[i] += 1
  else :
    stri[i] = 1

print(f"{stri}")


# 강사님 풀이
# 사용자로부터 문장을 입력 받습니다
sentence = input("문장을 입력하세요: ")

# 문장을 단어 단위로 분리합니다
words = sentence.split()

# 단어의 빈도수를 저장할 딕셔너리를 생성합니다.
word_frequency = {}

# 각 단어의 빈도수를 계산합니다.
for word in words:
  # 대소문자를 구분하지 않기 위해 단어를  소문자로 변환합니다.
  word = word.lower() # upper() 대문자로 변경

  # 딕셔너리에 단어가 이미 존재한다면 빈도수를 1증가시키고, 존재하지 않는다면 새로운 항목으로 추가합니다
  if word in word_frequency:
    word_frequency[word] += 1
  else :
    word_frequency[word] = 1


# 결과를 출력합니다
print(word_frequency)
```

#### 2진수, 8진수, 16진수의 데이터 표현 방법

- 우리가 프로그래밍을 하면서 변수에 어떤 값을 저장할 때 10진수는 일상적 표현을 그대로 대입한다.
- 예를들어 15를 저장하고 싶다면 int a= 15; 와 같이 15라는 숫자를 그대로 대입하면 된다.
- 하지만 2진수, 8진수, 16진수는 표현법이 따로 정해져있다.<br><br>  

- 2진수는 대입하고자 하는 값 앞에 0b를 붙인다. (숫자0과 영문자b)
- 8진수는 대입하고자 하는 값 앞에 0o 을 붙인다. (숫자0 영문자o)
- 16진수는 대입하고자 하는 값 앞에 0x 를 붙인다. (숫자0과 영문자x)

- 16진수로 7A 이다.

```python
# 리스트안에 딕셔너리로 정보 저장
user_data = [
    {'name':'John','age':28,'color':'Blue'}
    ,{'name':'Alice','age':24,'color':'Red'}
    ,{'name':'Maria','age':32,'color':'Green'}
    ]

# 1. 사용자 정보 입력
name = input("이름을 입력해주세요.")
age = int(input("나이를 입력해주세요."))
color = input("좋아하는 색상을 입력해주세요.")

# 2. 입력받은 사용자 정보를 딕셔너리로 만든 후 list형식인 user_data에 저장한다
user_data.append({'name':name, 'age':age, 'color':color})

print(f"\n전체 사용자 정보 : {user_data}")

total_age = 0

# 3. 이름이 John인 사용자 정보만 출력
for i in range(len(user_data)) :
  total_age += user_data[i]['age']
  if user_data[i]['name'] == 'John' :
    #print(user_data[i])
    print(f'\n이름 : {user_data[i]["name"]} / 나이 : {user_data[i]["age"]} / 좋아하는 색상 : {user_data[i]["color"]}')

# 4. 전체 사용자의 평균나이 구하기
average_age = total_age / len(user_data)

print(f"\n사용자 평균나이 : {average_age}")
```
