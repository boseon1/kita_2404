```python
import keyword

strings = ["var1","2things","variable_name","time!"]

def identifyKey(strings):

  count = 0
  # 키워드사용안됨
  for i in range(len(strings)):
    print(i)
    print(strings[i][0])

    # 넣은 변수와 키워드 비교 후 없으면 리턴
    if i not in keyword.kwlist:
      print(i)
    # 있으면
      break
      print('Not Available')

  # 첫 인덱스에 숫자난 공백 사용 불가
  # 첫 인덱스가 0부터 9인 숫자인 경우 제거
  for j in range(10):
    if strings[i][0] == str(j):
      print(strings[i][0])
      i -= 1
      # 해당 인덱스 제거
      strings.pop(i)

      print('ok')

  # 언더바 외 특수문자와 공백 사용 불가
  for k in strings:
    print(strings[i][k])
    count += 1
    print(k[i])
    print(strings[i][k])
    count += 1
    print(k[i])
    print(strings[k])
    # 각 리스트의 인덱스에 특수문자와 공백 있을 시 제거
    #if k[i] == '!':
    if i[j] == '@' or i[j] == '#' or i[j] == '%' or i[j] == '^' or i[j] == '*' or i[j] == ' ':
      strings.pop(k)

  # 결과 값 리턴
  return strings

print(identifyKey(identifiers))
```

#### sort 와 sorted 차이점

1. 메서드 vs 함수:
- sort(): 리스트 객체의 내장 메서드입니다. 즉, 리스트에서만 사용할 수 있습니다.
- sorted(): 내장 함수로, 어떤 반복 가능한(iterable) 객체에도 사용될 수 있습니다. 예를 들면 리스트, 튜플, 딕셔너리, 문자열 등에 사용할 수 있습니다.

2. 반환 값:
- sort(): 리스트를 원 위치에서(in-place) 정렬하고 None을 반환합니다. 따라서 원래의 리스트 자체가 변경됩니다.
- sorted(): 정렬된 새로운 리스트를 반환합니다. 원래의 객체는 변경되지 않습니다.

3. 유용성:
- sort(): 리스트에서만 작동하기 때문에 리스트만 정렬할 수 있습니다.
- sorted(): 다양한 객체를 정렬할 수 있으며 결과는 항상 리스트로 반환됩니다.

```python
# sort()
my_list = [3, 1, 2]
my_list.sort()
print(my_list)

# sorted()
my_tuple = (3, 1, 2)
new_list = sorted(my_tuple)
print(my_tuple)
print(new_list)
```

Task5_0425 주어진 정수 리스트와 회전 횟수 k에 대해 리스트를 오른쪽으로 k만큼 회전시킨 결과를 반환하세요. k가 리스트의 길이보다 클 수 있으며, 이 경우 k를 리스트 길이로 나눈 나머지만큼 실제 회전시키면 됩니다.

예를 들어, [1, 2, 3, 4, 5]와 k=2가 주어지면, 결과는 [4, 5, 1, 2, 3]이 되어야 합니다.

```python
list = [1, 2, 3, 4, 5]
k = 2

# 첫번째
list2 = list[-k::]
#print(list[-k::])
#print(list2)

# 두번째
list2.extend(list[:list.index(list2[0])])
#print(list.index(list2[0]))
#print(list[:3])

print(list2)

# ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
# k가 list길이보다 무조건 작을 경우
list = [1, 2, 3, 4, 5]
k = 4
list2 = list[-k::]
list2.extend(list[:list.index(list2[0])])
print(list2)
```

```python
# k가 list 길이보다 클수도 있을 경우
list = [1, 2, 3, 4, 5]
k = int(input())

#print(len(list))

if len(list) < k :
  #print(k%len(list))
  k = k%len(list)

# 먼저 시작 인덱스 부터 오른쪽에 있는 값을 list2에 담는다
list2 = list[-k::]

# 그 후 list2의 첫번째 값이 있는 위치까지의 list인덱스를 구해서 해당 값 앞에까지 list2에 담는다
list2.extend(list[:list.index(list2[0])])

print(list2)
```
