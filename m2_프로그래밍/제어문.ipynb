제어문(control statement)은 프로그램의 흐름을 제어하는데 사용

#### 조건문 (Conditional Statements):
- if: 특정 조건이 참(True)일 때 코드 블록을 실행
    - 조건은 비교 연산자(==, !=, <, >, <=, >=)와 논리 연산자(and, or, not)를 사용하여 표현할 수 있다
- elif: 이전의 if 또는 elif 조건이 거짓(False)일 때 다른 조건을 검사
    - elif는 'else if'의 축약형으로, 이전 if 또는 elif 문의 조건이 거짓(False)일 때 다른 조건을 검사
    - 여러 elif 문을 사용하여 다양한 조건을 검사할 수 있다
- else:
    - 이전의 if 또는 elif 문의 조건이 모두 거짓일 때 실행할 코드 블록을 지정

```python
x = 10
if x >5 :
  print("x는 5보다 큽니다.")
elif x == 5 :
  print("x는 5입니다.")
else :
  print("x는 5보다 작습니다")
```

```python
x = 10
if x >= 5:
  print("x는 5보다 크거나 같습니다")
else :
  print("x는 5보다 작습니다")
```

```python
# 각 조건이 서로 독립적일때 if문을 연속해서 사용
x = 10
y = 5

if x > 5:
  print("x는 5보다 큽니다.")

if y < 10 :
  print("y는 10보다 작습니다")

if x > y :
  print("x는 y보다 큽니다.")
```

```python
# Q. 이중 조건문을 사용하여 코드를 작성하세요.(조건 내용은 자유. 단, if문 if ~elif ~else 중첩하여 사용)

score = int(input("점수를 입력하세요 : "))

if score >= 90:
  if score >= 95:
    print("A+")
  else :
    print("A")
elif score >= 80 :
  if score >= 85 :
    print("B+")
  else :
    print("B")
elif score >= 70 :
  if score >= 75:
    print("C+")
  else :
    print("C")

```

```python
# Q. listdata = ['a','b',''c']에서 조건문을 이용해서 아래와 같이 출력하세요

# a가 listdata에 있습니다
# b가 listdata에 있습니다
# c가 listdata에 있습니다

listdata = ['a','b','c']

if 'a' in listdata :
  print('a가 listdata에 있습니다.')

if 'b' in listdata :
  print('b가 listdata에 있습니다.')

if 'c' in listdata :
  print('c가 listdata에 있습니다.')

print("\n")

for i in listdata :
  if i in listdata :
    print(f"{i}가 listdata에 있습니다")
```

```python
# Q. "수입(income)과 지출(spend)을 비교해서 수입이 많으면 저축증가, 반대이면 빚이 증가" 프로그램 작성

income = int(input("수입을 입력하세요."))
spend = int(input("지출을 입력하세요."))

if income > spend :
  print("저축증가")
elif income < spend :
  print("빚증가")
else :
  print("현상유지")
```

#### 반복문 (Loop Statements):
- for: 시퀀스(리스트, 튜플, 문자열 등)의 아이템을 순회하면서 코드 블록을 실행
- while: 특정 조건이 참인 동안 코드 블록을 반복하여 실행

```python
for i in range(5):
  print(i,end=' ')

print('\n')

count = 0

while count < 5 :
  print(count, end= ' ')
  count += 1
```

```python
# 리스트를 순회하는 예
for i in [1,2,3,4,5] :
  print(i ,end=' ')
print('\n')

# 문자열을 순회하는 예
for c in "hello" :
  print(c, end= ' ')
```

```python
# while 루프는 주어진 조건이 참(True)인 동안 코드 블록을 반복
# 조건이 거짓(False)이 되면 루프가 종료
count = 0
sum = 0
while count < 5 :
  sum += count
  print(sum, end= ' ')
  count += 1
```

```python
# Q. d에서 반복문을 사용하여 97 98 99을 출력하세요.
d = {'a' : 97, 'b':98, 'c' : 99}

for key,value in d.items() :
  print(f"key : {key}, value : {value}")

for key in d:
  print(d[key], end=' ')

d.values()
```
